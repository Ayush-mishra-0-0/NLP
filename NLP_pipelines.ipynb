### 7. MULTI-GENRE NATURAL LANGUAGE INFERENCE(MNLI)

```python
## HERE GIVING MODEL IS IMP FOR MNLI AS DEFAULT MODEL FOR TEXT CLASSIFICATION IS NOT FOR MNLI

mnli= pipeline(task="text-classification", model="roberta-large-mnli")
```

```python
mnli(
    "A man is black.\
    A black man is playing cricket"
)
```

```python
mnli(
    "Tanvi is a girl.\
    She loves maths"
)
```

```python
mnli(
    "Tanvi wife of ayush.\
    Tanvi will marry ayush."
)
```

### 8. Question Natural Language Inference(QNLI)

### 9. QQP(quora question pairs)

```python
model= "textattack/bert-base-uncased-QQP"
qqp= pipeline(task="text-classification",model=model)
```

```python
qqp(
    "what is the name of prime-minister of india?,\
    who is the prime-minister of india?"
)
```

```python
qqp(
     "what is the name of prime-minister of india?,\
    What is calculus?"
)
```

### 10. Table Question answering

```python
import pandas as pd
```

```python
tqa= pipeline(task="table-question-answering",model= "google/tapas-large-finetuned-wtq")
```

```python
data={
    "Actor":["Brad Pitt","Leonardo Di Caprio","George Clooney"],
    "Number of movies":["87","53","69"]
}
```

```python
df= pd.DataFrame.from_dict(data)
```

```python
df.head()
```

```python
q="How may movies are of brad pitt?"
x=tqa(
    table=df,
    query=q
)
```

```python
x
```
